```python
#zeros in columns like Glucose, BloodPressure, SkinThickness, Insulin, and BMI do not represent valid values (as they cannot be zero for a person). So, we'll handle them by replacing zeros with NaN and then fill them with the median value of each column.
df[["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]]=df[["Glucose", "BloodPressure", "SkinThickness", "Insulin", "BMI"]].replace(0,np.nan)
```

```python
# shuffle data
df = df.sample(frac=1)
df = df.reset_index(drop=True)  # reet index after shuffling
```

```python
df.head()
```

```python
# Split data into features and target

X = df.drop("Outcome",axis=1)
Y = df["Outcome"]
```

```python
X.shape, Y.shape
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score
```

```python
# Split the data into training and testing sets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
param_grid = {
    'n_estimators': [25, 50, 75, 100],       # Balanced number of trees for quick training and sufficient diversity.
    'max_depth': [3, 5, 7, 10],       # Shallow to fully grown trees for flexibility.
    'max_features': ['sqrt', 'log2'],   # Subsets of features to prevent overfitting.
    'min_samples_split': [2, 5, 10],    # Fine-to-coarse splits based on dataset size.
    'class_weight': ['balanced']        # Adjust weights to handle the slight imbalance.
}
```

```python
# Initialize the Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42)
```

```python
# Set up GridSearchCV
grid_search = GridSearchCV(estimator=rf_model, param_grid=param_grid, cv=3, scoring='accuracy', verbose=1)
```

```python
# Perform Grid Search
grid_search.fit(X_train, Y_train)
```

```python
# Extract results
results = pd.DataFrame(grid_search.cv_results_)
```

```python
# Select relevant columns for meta-dataset
meta_dataset = results[['param_n_estimators', 'param_max_depth', 'param_max_features', 'param_min_samples_split',
                        'mean_test_score']]
```

```python
# Rename columns for clarity
meta_dataset.rename(columns={
    'param_n_estimators': 'n_estimators',
    'param_max_depth': 'max_depth',
    'param_max_features': 'max_features',
    'param_min_samples_split': 'min_samples_split',
    'mean_test_score': 'accuracy'
}, inplace=True)
```

```python
# Add meta-features of the dataset
meta_features = {
    'num_features': X.shape[1],
    'num_samples': X.shape[0],
    'class_imbalance_ratio': Y.value_counts().max() / Y.value_counts().min(),
    'num_classes': len(Y.unique())
}
```

```python
for feature, value in meta_features.items():
    meta_dataset[feature] = value
```

```python
# Display the meta-dataset
print(meta_dataset.head())
```

```python
# Save the meta-dataset to a CSV file for later use
meta_dataset.to_csv('meta_dataset_diabetes.csv', index=False)
```
